### Filter crop_recommendation.csv to MergeFileCrop.cv
### Filter fertilizer.csv to MergerFileFert.csv

```python
import pandas as pd
```

```python
# Reading the data 

crop_data_path = '../Datasets/Crop_recommendation.csv'
fertilizer_data_path = '../Datasets/Fertilizer.csv'

crop = pd.read_csv(crop_data_path)
fert = pd.read_csv(fertilizer_data_path)
```

```python
crop.head()
```

```python
fert.head()
```

```python
# Function for lowering the cases
def change_case(i):
    i = i.replace(" ", "")
    i = i.lower()
    return i
```

```python
fert['Crop'] = fert['Crop'].apply(change_case)
crop['label'] = crop['label'].apply(change_case)
```

```python
#make some changes in ferttilizer dataset

fert['Crop'] = fert['Crop'].replace('mungbeans','mungbean')
fert['Crop'] = fert['Crop'].replace('lentils(masoordal)','lentil')
fert['Crop'] = fert['Crop'].replace('pigeonpeas(toordal)','pigeonpeas')
fert['Crop'] = fert['Crop'].replace('mothbean(matki)','mothbeans')
fert['Crop'] = fert['Crop'].replace('chickpeas(channa)','chickpea')
```

```python
crop_names_from_fert = fert['Crop'].unique()
crop_names_from_fert
```

```python
extract_labels = []
for i in crop_names_from_fert:
    if i in crop_names:
        extract_labels.append(i)
```

```python
# using extract labesl on crop to get all the data related to those labels
new_crop = pd.DataFrame(columns = crop.columns)
new_fert = pd.DataFrame(columns = fert.columns)
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'crop' and you want to create a new DataFrame 'new_crop'
new_crop = pd.DataFrame()

# Loop through extract_labels and concatenate rows to new_crop
for label in extract_labels:
    subset = crop[crop['label'] == label]
    new_crop = pd.concat([new_crop, subset], ignore_index=True)
```

```python
# Assuming you have a DataFrame named 'fert' and you want to create a new DataFrame 'new_fert'
new_fert = pd.DataFrame()

# Loop through extract_labels and append the first row to new_fert for each label
for label in extract_labels:
    subset = fert[fert['Crop'] == label]
    if not subset.empty:  # Check if the subset is not empty
        new_fert = pd.concat([new_fert, subset.iloc[0]])
```
